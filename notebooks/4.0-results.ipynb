# Results/plots

This notebook contains score plots and code for result analysis

```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append("../src")

import utils.misc as u

from altair import *
```

```python
EXP_NAME = "collaborative"

def plot_target(target):
    """
    Plot scores for the target given
    """
    
    df = u.get_scores("../results", EXP_NAME, str(target))
    # Remove all region
    df = df[df["region"] != "all"]
    df["score"] = -df["score"]
    
    chart = Chart(df).mark_point().encode(
        x="score", y=Y(
            "model",
            sort=SortField(field="score", op="mean", order="descending")
        ), color="region"
    )
    return chart
```
